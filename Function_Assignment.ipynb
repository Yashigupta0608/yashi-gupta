1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.




```python
def sum_of_evens(numbers):
  return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6, 7, 8,]
print(sum_of_evens(nums))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        fib_num = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib_num

fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
import os

def read_file_lines(file_path):
    if not os.path.exists(file_path):
        print(f"Error: File '{file_path}' not found.")
        return  # Stop execution if the file does not exist

    with open(file_path, "r") as file:
        for line in file:
            yield line.strip()

file_path = "example.txt"  # Ensure this file exists
for line in read_file_lines(file_path):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 20, 30, 40, 100]

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, s))

text = "Hello, World!"
result = remove_vowels(text)
print(result)
```

product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
def process_orders(orders):

    order_data = [(order[0], order[3], order[4]) for order in orders]

    result = list(map(
        lambda x: (
            x[0],
            x[1] * x[2] + 10 if x[1] * x[2] < 100 else x[1] * x[2]
        ),
        order_data
    ))

    return result

orders = [
    [34587, "Learning Python, Mark Lutz", "Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", "Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", "Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", "Bernd Klein", 3, 24.99]
]

results = process_orders(orders)
for order_number, total in results:
    print(f"Order {order_number}: {total:.2f}€")
```
